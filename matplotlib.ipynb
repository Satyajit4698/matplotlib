```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)
y2=np.cos(2*np.pi*x2)*np.exp(-x1)

plt.subplot(2,1,1)
plt.plot(x1,y1,'-o')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('cos wave plot')
plt.grid()

plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('amp y2')
plt.grid()
plt.show


```

# Histogram Graph

```python
import matplotlib.pyplot as plt
ages=[2,3,54,32,12,34,56,76,65,45,42,65,47,46,43,41,49,42]
range=(0,100)
bins=10
plt.hist(ages,bins,range,color='red',histtype='bar',rwidth=2.0)
plt.xlabel('ages')
plt.ylabel('bins')
plt.title('histogram graph')
plt.grid()
plt.show
```

# Piechart

```python
activity=['red','slip','eat','travel']
slices=[4,9,7,5]
color=['r','g','b','y']
plt.pie(slices,labels=activity,colors=color,shadow='true',startangle=90,explode=(0.2,0,0,0),autopct='%1.3f%%')
plt.title('pie')
plt.legend()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D


# create cluster builder
def cluster(center, radius=10, n=50):
    c = np.asarray(center)
    xx, yy, zz = np.random.uniform(c-radius, c+radius, size=(n, 3)).T
    return xx, yy, zz


# create random cluster
xx1, yy1, zz1 = cluster((25, 15, 5))
xx2, yy2, zz2 = cluster((5, 5, 20))

# create a figure
fig = plt.figure()
# initialise 3D Axes
ax = Axes3D(fig)
ax.grid(False)


# create the initialiser with our two clusters
def init():
    ax.scatter(xx1, yy1, zz1, marker='o', s=40, c='#08C6AB', alpha=0.6)
    ax.scatter(xx2, yy2, zz2, marker='o', s=40, c='#37465B', alpha=0.6)
    return fig,


# create animate function, this will adjust the view one step at a time
def animate(i):
    ax.view_init(elev=10.0, azim=i)
    return fig,


# create the animated plot
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=360, interval=20, blit=True)
# save as a GIF
anim.save('clusters.gif', fps=30, writer='imagemagick')
```
